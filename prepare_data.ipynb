```python
import pandas as pd

data = pd.read_csv("AAPL.csv")

close = data["Close"].values

close = close[2:]

close = close.astype("float")

print(close[:10])
print()
print(close[10:20])

```

```python
X = []
y = []

sliding_window_size = 10

print(len(close))
```

```python
for i in range(sliding_window_size, len(close)):
    y.append(close[i])
    X.append(close[i-sliding_window_size:i])

print(X[:5])
print()
print(y[:5])
```

```python
print(len(X))
print(len(y))   
```
